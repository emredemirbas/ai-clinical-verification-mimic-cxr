```python
import pandas as pd

# Merge the results and ground truth dataframes
merged = pd.merge(results_df, ground_truth_df, on=['subject_id', 'study_id'], suffixes=('_results', '_ground_truth'))

# Initialize a list to store results
results = []

# Columns to compare (excluding 'subject_id' and 'study_id')
columns_to_compare = [col for col in results_df.columns if col not in ['subject_id', 'study_id']]

# Iterate through each column to calculate metrics
for col in columns_to_compare:
    # Get the results and ground truth for the current column
    y_pred = merged[f"{col}_results"]
    y_true = merged[f"{col}_ground_truth"]

    # Calculate total matches for accuracy
    matches = (y_pred == y_true).sum()
    total = len(y_true)
    accuracy = (matches / total) * 100

    # Calculate precision, recall, and F1-score for each class
    classes = ['Yes', 'No', 'Maybe', 'Undefined']
    tp, fp, fn = 0, 0, 0
    for cls in classes:
        tp += ((y_pred == cls) & (y_true == cls)).sum()  # True Positives
        fp += ((y_pred == cls) & (y_true != cls)).sum()  # False Positives
        fn += ((y_pred != cls) & (y_true == cls)).sum()  # False Negatives

    # Calculate Precision, Recall, and F1-Score
    precision = (tp / (tp + fp) * 100) if (tp + fp) > 0 else 0
    recall = (tp / (tp + fn) * 100) if (tp + fn) > 0 else 0
    f1_score = (2 * precision * recall / (precision + recall)) if (precision + recall) > 0 else 0

    # Append results for the current column
    results.append({
        'Category': col,
        'Total': total,
        'Matches': matches,
        'Accuracy': f"{accuracy:.2f}%",
        'Precision': f"{precision:.2f}%",
        'Recall': f"{recall:.2f}%",
        'F1-Score': f"{f1_score:.2f}%"
    })

# Create a DataFrame to display the results
metrics_df = pd.DataFrame(results)

# Display the metrics DataFrame
metrics_df
```

## Sadece Yes ve No'lar üzerinden değerlendirme

```python
# Merge the results and ground truth dataframes
merged = pd.merge(results_df, ground_truth_df, on=['subject_id', 'study_id'], suffixes=('_results', '_ground_truth'))

# Initialize a list to store results
results = []

# Columns to compare (excluding 'subject_id' and 'study_id')
columns_to_compare = [col for col in results_df.columns if col not in ['subject_id', 'study_id']]

# Iterate through each column to calculate metrics
for col in columns_to_compare:
    # Filter rows where both results and ground truth are either 'Yes' or 'No'
    filtered = merged[
        (merged[f"{col}_results"].isin(['Yes', 'No'])) &
        (merged[f"{col}_ground_truth"].isin(['Yes', 'No']))
    ]
    
    # Total number of valid rows for this column
    total_valid = len(filtered)
    
    # Calculate True Positives (TP)
    tp = ((filtered[f"{col}_results"] == 'Yes') & (filtered[f"{col}_ground_truth"] == 'Yes')).sum()
    
    # Calculate False Positives (FP)
    fp = ((filtered[f"{col}_results"] == 'Yes') & (filtered[f"{col}_ground_truth"] == 'No')).sum()
    
    # Calculate False Negatives (FN)
    fn = ((filtered[f"{col}_results"] == 'No') & (filtered[f"{col}_ground_truth"] == 'Yes')).sum()
    
    # Calculate Precision
    precision = (tp / (tp + fp) * 100) if (tp + fp) > 0 else 0
    
    # Calculate Recall
    recall = (tp / (tp + fn) * 100) if (tp + fn) > 0 else 0
    
    # Calculate F1-Score
    f1_score = (2 * precision * recall / (precision + recall)) if (precision + recall) > 0 else 0
    
    # Calculate Matches (for Accuracy)
    matches = (filtered[f"{col}_results"] == filtered[f"{col}_ground_truth"]).sum()
    
    # Calculate Accuracy
    accuracy = (matches / total_valid * 100) if total_valid > 0 else 0
    
    # Append results
    results.append({
        'Category': col,
        'Total Valid': total_valid,
        'Matches': matches,
        'Accuracy': f"{accuracy:.2f}%",
        'Precision': f"{precision:.2f}%",
        'Recall': f"{recall:.2f}%",
        'F1-Score': f"{f1_score:.2f}%"
    })

# Create a DataFrame to display the results
metrics_df = pd.DataFrame(results)

# Display the metrics DataFrame
metrics_df
```
